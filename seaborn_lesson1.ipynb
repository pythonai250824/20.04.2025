```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

plt.figure(figsize=(8,4)) 

df = pd.read_csv("dm_office_sales.csv")

#sns.scatterplot(data=df, x='salary', y='sales',hue="salary")
#sns.scatterplot(data=df, x='salary', y='sales',hue="work experience", palette='dark')
sns.scatterplot(data=df, x='salary', y='sales',s=120, hue="training level", palette='Spectral', alpha=0.7, marker='o')

plt.tight_layout()
plt.legend(loc=(1.1, 0))
plt.show()
```

```python
df = pd.read_csv("dm_office_sales.csv")

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

plt.figure(figsize=(12,4)) 

#sns.countplot(data=df, x='training level')

df['training level'] = df['training level'].astype(str)

sns.countplot(data=df, x='level of education', hue="training level", palette='Spectral')
```

```python


df = pd.read_csv("dm_office_sales.csv")

plt.figure(figsize=(10, 4))

df
display(df.groupby('level of education').agg({'salary':['min', 'max', 'mean', 'std']}))

# display(df.groupby('level of education').max('level of education'))  # in order to to max need manually in this barplot

#sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, errorbar='sd')

sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, errorbar='sd')
```
